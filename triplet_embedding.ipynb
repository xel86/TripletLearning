```
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```
def triplet_probability(triplets, alpha):
	#probablity that i is more similar to j than l
  try:
    X_i, X_j, X_l = torch.FloatTensor(triplets[0]), torch.FloatTensor(triplets[1]), torch.FloatTensor(triplets[2]) 
    pij = torch.pow((1 + (torch.pow(torch.abs(X_i - X_j), 2) / alpha)), -(alpha + 1) / 2)
    pil = torch.pow((1 + (torch.pow(torch.abs(X_i - X_l), 2) / alpha)), -(alpha + 1) / 2)
    p = pij / (pij + pil)
  except:
    raise Exception("Input format for triplet probablity is array of [Xi, Xj, Xl] with X's having same shape, and alpha ([[1,2,3], [3,2,1], [4,5,6]], alpha)")
  return p

#def triplet_probability(dij, dil, alpha):
```
